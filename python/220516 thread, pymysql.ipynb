```python
import time
import multiprocessing

def count_down(x):
    while True:
        if x == 0:
            break
            
        print("카운트 다운 ... %d" % x)
        x -= 1
        time.sleep(1)
        
if __name__ == '__main__':
    p = multiprocessing.Process(target=count_down, args=(5,))
    print(p)
    p.start()
    p.join()
```

```python
import threading
import os

def info():
    print('module name:', __name__)
    print('process id: ', os.getpid())
    print('thread id:', threading.get_ident())
    print('이런식으로 많이 집어넣으면 context switching도 직접 확인 가능')
    #제어권이 왔다갔다하는거 확인 가능
    #뷰에서 어싱크랑 웨이트를 사용해서 동작이 끝나는걸 기다려줬었음

def print_hello(): 
    print("안녕")
    info()
    time.sleep(1)
    print("잠자다 일어나기\n")
    
if __name__ == '__main__':
    start = time.perf_counter()
    threads = []
    
    for _ in range(5):
        t = threading.Thread(target=print_hello)
        t.start()
        threads.append(t)
        
    for thread in threads:
        thread.join()

    finish = time.perf_counter()
    
    print('종료: ', str(round(finish - start, 2)) + "초")
```

```python
lock = threading.Lock()
x = 0 

def increment_global(): #크리티컬 섹션 문제: 실행을 보장할 수 없어서 변경되기 전 값을 가져옴 -> Lock 추가
    global x 
    x += 1
    
def thread_processing():
    lock.acquire() #작업의 시작 보장
    for _ in range(1000000):
        increment_global()
    lock.release() #작업의 끝 보장
    #제어권을 먼저 갖는 쓰레드가 동작하는 동안 다른 쓰레드는 동작하지 못함
        
def thread_main():
    global x 
    x = 0
    
    t1 = threading.Thread(target = thread_processing)
    t2 = threading.Thread(target = thread_processing)
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
for i in range(10):
    thread_main()
    print("{0}번째 반복 이후 x = {1}".format(i, x))
```

```python
import pymysql
```

```python
#작업하는 위치에 mysql파일 생성한 후 실행
MYSQL_USER_DATA_SAVED_FILE="mysql/userinfo" 
```

```python
import pickle

mysql_user_info = dict({
    'user_id': 'eddi',
    'password': 'eddi@123'
})

f = open(MYSQL_USER_DATA_SAVED_FILE, 'wb')
pickle.dump(mysql_user_info, f) #특정 객체를 저장하여 객체 자체를 불러와서 사용하기 위함
f.close()

del mysql_user_info
```

```python
import pickle

f = open(MYSQL_USER_DATA_SAVED_FILE, 'rb')
mysql_user_info = pickle.load(f)
f.close()
```

```python
#터미널에서 스키마(pydb) 생성하고 실행 -> root계정으로 스키마 생성하고 권한까지 줬음
db = pymysql.connect(
    host = '127.0.0.1',
    port = 3306,
    user = mysql_user_info['user_id'],
    passwd = mysql_user_info['password'],
    db = 'pydb'
) #유저인포에 저장된 로그인 정보로 pydb에 접근하겠다
cursor = db.cursor() #디비제어에 필요
print(db)
```

```python
sql = """
    create table pydb_test(
    id int unsigned not null auto_increment, 
    name varchar(20) not null, 
    price int not null,
    primary key(id)
    )
"""

cursor.execute(sql) #쿼리실행
db.commit()
db.close()
print('테이블 생성 성공')
```
